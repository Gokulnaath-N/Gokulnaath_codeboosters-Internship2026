```python
# Average total score of srudents with attendence above 90 %

avg_total_score = df[df['attendance_percentage']>90]['total_score'].mean().round(2)

avg_total_score
```

## Score Analysis

```python
#Finding highest math score

high_math_score=df['math_score'].max()
print("Highest math score : ",high_math_score)
print()

#Print the lowest math score
low_math_score=df['math_score'].min()
print("Lowest math score :",low_math_score)
print()

#Print the average math score
avg_math_score=df['math_score'].mean().round(2)
print("Average math score :",avg_math_score)
print()

#Print the average programming score
avg_prog_score=df['programming_score'].mean()
print("Average Programming score : ",avg_prog_score)
print()


#Average attendence score
avg_attend_score = df['attendance_percentage'].mean()
print("Average Attendence percentage : ",avg_attend_score)
print()



```

## Calculating and display average total score grouped by gender

```python
df.groupby('gender')['total_score'].mean()
```

## To display the top performers name, department and total score

```python
#To display the top 3 performers name, department and total score

top_performers = df.nlargest(3, 'total_score')
top_performers[['name','department','total_score']]
```
